```python
import matplotlib.pyplot as plt 
import numpy as np

t_a = 100
iteration = 90 

freq = 1e9 # Hz GHz = 1e9
bandwidth = 500*1024 # Hz GHz = 1e9

Kb = 1024
Mb = 1024*1024
Gb = 1024*1024*1024

proposed_cmf = np.load("Proposed_"+str(freq)+"_"+str(bandwidth)+"_cmf.npy")
proposed_ho_cmf = np.load("Proposed_"+str(freq)+"_"+str(bandwidth)+"_ho_cmf.npy")
for i in range(len(proposed_cmf)):
    proposed_cmf[i] = proposed_cmf[i] /(i+1)


random_cmf = np.load("Random_"+str(freq)+"_"+str(bandwidth)+"_cmf.npy")
random_ho_cmf = np.load("Random_"+str(freq)+"_"+str(bandwidth)+"_ho_cmf.npy")
for i in range(len(random_cmf)):
    random_cmf[i] = random_cmf[i] /(i+1)


RSSI_cmf = np.load("RSSI_"+str(freq)+"_"+str(bandwidth)+"_cmf.npy")
RSSI_ho_cmf = np.load("RSSI_"+str(freq)+"_"+str(bandwidth)+"_ho_cmf.npy")
for i in range(len(RSSI_cmf)):
    RSSI_cmf[i] = RSSI_cmf[i] /(i+1)

AB_cmf = np.load("AB_"+str(freq)+"_"+str(bandwidth)+"_cmf.npy")
AB_ho_cmf = np.load("AB_"+str(freq)+"_"+str(bandwidth)+"_ho_cmf.npy")
for i in range(len(AB_cmf)):
    AB_cmf[i] = AB_cmf[i] /(i+1)

LONGEST_cmf = np.load("LONGEST_"+str(freq)+"_"+str(bandwidth)+"_cmf.npy")
LONGEST_ho_cmf = np.load("LONGEST_"+str(freq)+"_"+str(bandwidth)+"_ho_cmf.npy")
for i in range(len(LONGEST_cmf)):
    LONGEST_cmf[i] = LONGEST_cmf[i] /(i+1)

methods = ['proposed', 'random', 'RSSI']
```

```python
plt.plot(proposed_cmf[10:] / (60*Kb), label="PROPOSED", color='m')
plt.plot(random_cmf[10:] / (60*Kb), label="RANDOM", color='g')
plt.plot(RSSI_cmf[10:] / (60*Kb), label="RSSI", color='b')
plt.plot(LONGEST_cmf[10:] / (60*Kb), label="LONGEST", color='r')
plt.plot(AB_cmf[10:] / (60*Kb), label="GREEDY", color='y')

plt.grid(True)

plt.xlabel('Time [1m]')
plt.ylabel('Channel Capacity [Kbps]')
plt.title('Channel Capacity Comparison')
plt.legend()
plt.show()
```

```python
plt.plot(proposed_ho_cmf, label="PROPOSED")
plt.plot(random_ho_cmf, label="RANDOM")
plt.plot(RSSI_ho_cmf, label="RSSI")
plt.plot(LONGEST_ho_cmf, label="LONGEST")
plt.plot(AB_ho_cmf, label="GREEDY")
plt.grid(True)

plt.xlabel('Time [1m]')
plt.ylabel('Handover')
plt.title('Cumulative Handover Comparison')
plt.legend()
plt.show()
```

```python
proposed_sat_receive_avg_data = (proposed_cmf[-1]*180) / (proposed_ho_cmf[-1]* Mb)
random_sat_received_avg_data = (random_cmf[-1]*180) / (random_ho_cmf[-1]* Mb)
rssi_sat_received_avg_data = (RSSI_cmf[-1]*180) / (RSSI_ho_cmf[-1] * Mb)
longest_sat_received_avg_data = (LONGEST_cmf[-1]*180) / (LONGEST_ho_cmf[-1] * Mb)
ab_sat_received_avg_data = (AB_cmf[-1]*180) / (AB_ho_cmf[-1] * Mb)
methods = ['PROPOSED', 'RANDOM', 'RSSI', 'LONGEST', 'GREEDY']

x = np.arange(5)
plt.bar(x, (proposed_sat_receive_avg_data, random_sat_received_avg_data, rssi_sat_received_avg_data, longest_sat_received_avg_data, ab_sat_received_avg_data), color=['r', 'g', 'b', 'c', 'purple'])
plt.xticks(x, methods)
plt.xlabel('Methods')
plt.ylabel('Data Amount [Mb]')
plt.title("Data throughput on single Satellite")

plt.plot()
```
